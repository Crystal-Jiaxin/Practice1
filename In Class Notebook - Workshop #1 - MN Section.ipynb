```python
# Change Directory to Access Data Files #

os.chdir("/content/drive/MyDrive/Teaching/2022/DSO 574/Python Workshops/Data Files/AirBnB")

data = pd.read_csv("./LA2020listings.csv")

SM = data.loc[      data['neighbourhood']  == 'Santa Monica'                ]
LB = data.loc[      data['neighbourhood']  == 'Long Beach'                ]

SM['price'].mean()
LB['price'].mean()
```

```python
SM['AvePrice'] = SM['price'].mean()
```

```python
SM['DifferenceFromAverage'] = SM['price'] - SM['AvePrice']
```

```python
SM['DifferenceFromAverage'].max()
```

```python
LB['AvePrice'] = LB['price'].mean()
LB['DifferenceFromAverage'] = LB['price'] - LB['AvePrice']
LB['DifferenceFromAverage'].max()
```

```python
SM_AND_LB = pd.concat([SM, LB])
```
